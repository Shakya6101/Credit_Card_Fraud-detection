```python
import numpy as np
import pandas as pd
import matplotlib as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

 Loding data

```python
credit_card_data = pd.read_csv("creditcard.csv")
credit_card_data.head(5)
```

```python
credit_card_data.tail()
```

```python
#dataset informations
credit_card_data.info()
```

```python
#checking the numbers of valuse in each column
credit_card_data.isnull().sum()
```

```python
#distribution of legit transactions & fraudulent transactions
credit_card_data['Class'].value_counts()
```

This dataset is highly unbalanced
0 ---> legit
1 --->fraud

```python
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#statistical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#compare valuse for both transactions
credit_card_data.groupby('Class').mean()
```

```python
# build a sample dataset containing similar distribution of normal transactions and fraudulent transactions
# number of fraudulent transactions ---> 492
```

```python
legit_sample = legit.sample(n = 492)
```

Concatenating two dataframes

```python
new_dataset = pd.concat([legit_sample,fraud],axis = 0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

spliting th data into features & Targets


```python
X = new_dataset.drop(columns='Class',axis=1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Model Training 

Loigistic Regression

```python
model = LogisticRegression(max_iter=2000, solver='liblinear', multi_class='ovr')
```

```python
#Training the Logistic regression Model with Training Data
model.fit(X_train,Y_train)
```

model evaluation

```python
#accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
```

```python
print("Accuracy on Training data: ",training_data_accuracy)
```

```python
#accuracy on train data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction,Y_test)
```

```python
print("Accuracy on Test Data: ",test_data_accuracy)
```
